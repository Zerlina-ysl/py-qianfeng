```python
import time,random
from selenium import webdriver


class App():
    def __init__(self):
        self.host="http://www.douyu.com/"
        self.room=input('请输入房间号')#6415143
        self.user='18336456921'
        self.password='Demonhunter521'
        self.bow=webdriver.Chrome()
        
    #模拟登陆
    def login(self):
        self.bow.get(url=self.host+self.room)

        # 登陆交互
        time.sleep(7)
        loginBtn = self.bow.find_element_by_class_name('MuteStatus-isLog')
        time.sleep(5)
        loginBtn.click()
        time.sleep(18)
#登陆弹窗是一个子页面 所以要切换页面
        self.bow.switch_to.frame('login-passport-frame')
        time.sleep(1)
        self.bow.find_element_by_class_name('inputLoginBtn').click()
        time.sleep(2)
        phonenum = self.bow.find_element_by_xpath('//*[@id="loginbox"]/div[3]/div[2]/div/form/div[1]/div/input')

        phonenum.send_keys(self.user)
        password=self.bow.find_element_by_xpath('//*[@id="loginbox"]/div[3]/div[2]/div/form/div[3]/input[1]')
        password.send_keys(self.password)
#         self.bow.find_element_by_xpath('//*[@id="996h9juczu4"]/div/div[2]/div[1]/div[3]').click()
        time.sleep(12)
        self.bow.find_element_by_xpath('//*[@id="loginbox"]/div[3]/div[2]/div/form/div[6]/input').click()
        time.sleep(3)
        
    
    
    
    
    # 切回默认页面
        self.bow.switch_to.default_content()
    # 发送弹幕
    def send(self):
        danmu_list=[
            '支持主播',
            '主播加油',
            '主播好牛',
            '学到了'
        ]
        time.sleep(10)
        for i in range(0,10):
            time.sleep(25)
            self.bow.find_element_by_xpath('//*[@id="layout-Player-aside"]/div[2]/div/div[2]/div[3]/div[1]').send_keys('danmu_list[random.randint(0,len(danmu_list)-1)]')
            self.bow.find_element_by_xpath('//*[@id="layout-Player-aside"]/div[2]/div/div[2]/div[3]/div[3]').click()
        else:
            time.sleep(2)
            # 退出浏览器
            self.bow.quit()
    
    def main(self):
        self.login()
        self.send()
        
    
    
    
app = App()
app.main()
```

```python
from selenium import webdriver
import time,json

bow = webdriver.Chrome()
bow.get("http://www.douyu.com/6415143")
# 删除原本的cookie信息
bow.delete_all_cookies()
file=open('./cookies.txt','r')
# 使用json读取cookie py可以直接解析
cookies = json.load(file)
for cookie in cookies:
    # 添加cookie信息
    bow.add_cookie(cookie)
    
# 刷新
bow.refresh()
# 网页验证cookie 查看是否免登陆


# 发送弹幕
danmu_list=[
            '支持主播',
            '主播加油',
            '主播好牛',
            '学到了'
        ]
time.sleep(10)
for i in range(0,10):
    time.sleep(10)
    bow.find_element_by_xpath('//*[@id="layout-Player-aside"]/div[2]/div/div[2]/div[3]/textarea').send_keys(danmu_list[random.randint(0,len(danmu_list)-1)])
    bow.find_element_by_xpath('//*[@id="layout-Player-aside"]/div[2]/div/div[2]/div[3]/div[2]').click()
else:
    time.sleep(2)
            # 退出浏览器
    bow.quit()
    
```
