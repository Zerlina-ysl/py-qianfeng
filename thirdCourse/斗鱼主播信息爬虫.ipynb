```python
import selenium.webdriver,time,csv
# 需要将信息存储在csv

class App():
    def __init__(self):
        self.url = 'https://www.douyu.com/g_wzry'
        self.bow = selenium.webdriver.Chrome()
        
    def getHtml(self):
        # 1.等待加载信息
        time.sleep(8)
        # 2. 持续下拉滚动条 触发所有信息加载完全
        for i in range(2,90):
            # 使用脚本
            self.bow.execute_script("document.documentElement.scrollTop={}".format(i*100))
        else:
            time.sleep(15)
        # 3. 查询元素
        rooms = []
        li_list = self.bow.find_elements_by_xpath('//ul[@class="layout-Cover-list"]/li')
        print("内容长度：",len(li_list))
        for li in li_list:
            room = {}
            try:
            # 4.1 获取房间图片
            
                room['room_img']=li.find_element_by_xpath('.//img[@class="DyImg-content is-normal "]').get_attribute('src')

            # 4.2 获取房间标题
                room['room_title']=li.find_element_by_xpath('.//h3[@class="DyListCover-intro"]').text
            # 4.3 获取房间类型
                room['room_type']=li.find_element_by_xpath('.//span[@class="DyListCover-zone"]').text
            # 4.4 获取房间主播名称
                room['room_name']=li.find_element_by_xpath('.//div[@class="DyListCover-userName"]').text
            # 4.5 获取房间热度
                room['room_hot']=li.find_element_by_xpath('.//span[@class="DyListCover-hot"]').text
            
            # 5. 存储
                rooms.append(room)
            except:
                continue
#             print(room)
        else:
            # 6. 获取下一页数据 并判断是否存在
                # 最后一页的class类型会多加一个 dy-Pagination-disabled类型
            next_btn = li.find_elements_by_xpath('//li[@class=" dy-Pagination-next"]')
            # 确保下一页可以点击
            next_btn = next_btn[0] if len(next_btn)>0 else None
        return rooms,next_btn
            
            
        
        
    def parseHtml(self):
        pass
    def save(self,rooms):
        file = open(file='./dyRoomInfo.csv'.format(time.time()),mode='w',encoding="utf-8",newline='')
        file_csv=csv.writer(file)
        file_csv.writerow(rooms[0].keys())# 将下标作为表头
        file_csv.writerows([(room.values())for room in rooms])
        # 清空缓存
        file.flush()
        file.close()
        print("success,",file)
        
        
    def main(self):
            # 1. 设置浏览器属性
        # 设置宽高
        #self.bow.set_window_size()
        # 设置窗口最大化
        self.bow.maximize_window()
            # 2. 发起地址请求
        self.bow.get(url=self.url)
            # 3. 第一次调用：获取第一页数据
        rooms,next_btn = self.getHtml()
        self.save(rooms)
        # 7. 跳转下一页
        while next_btn!=None:
            next_btn.click()
            rooms,next_btn = self.getHtml()
            self.save(rooms)
        else:
            self.bow.quit()
app = App()
app.main()
```
